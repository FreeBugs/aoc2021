## Part 1

```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.rcParams.update({
    "lines.color": "white",
    "patch.edgecolor": "white",
    "text.color": "black",
    "axes.facecolor": "white",
    "axes.edgecolor": "lightgray",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "grid.color": "lightgray",
    "figure.facecolor": "black",
    "figure.edgecolor": "black",
    "savefig.facecolor": "black",
    "savefig.edgecolor": "black"})

with open('d06.txt', 'r') as f:
    content = f.readlines()

fish0 = [int(t) for t in content[0].strip('\n').split(',')]
```

```python
fish = np.array(fish0)
fish_to_day = [(fish == d).sum() for d in range(0, 9)]
print(fish_to_day)
days_to_process = 80
for day in range(0, days_to_process):
    zero_fish = fish_to_day[0]
    fish_to_day = fish_to_day[1:] + [zero_fish]
    fish_to_day[6] += zero_fish
    print(fish_to_day)

print(sum(fish_to_day))
```

## Part 2

```python
fish = np.array(fish0)
fish_to_day = [(fish == d).sum() for d in range(0, 9)]
print(fish_to_day)
days_to_process = 256
fishcount = []
for day in range(0, days_to_process):
    zero_fish = fish_to_day[0]
    fish_to_day = fish_to_day[1:] + [zero_fish]
    fish_to_day[6] += zero_fish
    fishcount.append(sum(fish_to_day))
    print(fish_to_day)

print(sum(fish_to_day))
```

```python
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(1, 1, 1)
ax.set_facecolor((.1,.1,.1))
p = ax.plot(fishcount, c='yellow')
```
