## Part 1

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import networkx as nx

with open('d14.txt', 'r') as f:
    content = f.readlines()

template = content[0]
data = [c.split(' -> ') for c in [d.strip('\n') for d in content[2:]]]
data = {c[0]: [c[1], c[0][0]+c[1]+c[0][1]] for c in data}
chars = list(set(d[0] for d in data.values()))
data
```

```python
pd.DataFrame(columns=list(data.keys()), data=[[-1 if p == d else 1 if d in data[p][1] else 0 for d in data.keys()] for p in data.keys()], index=list(data.keys()))
```

```python
pairs = np.asarray([template.count(d) for d in data.keys()])
pairs
```

```python
pairs = np.transpose(np.asarray([template.count(d) for d in data.keys()]))
char_counts = np.asarray([template.count(d) for d in chars])
increments = [[0 if p==d and d in data[p][1] else -1 if p == d else 1 if d in data[p][1] else 0 for d in data.keys()] for p in data.keys()]
char_increments = [[1 if d == data[p][0] else 0 for d in chars] for p in data.keys()]
for i in range(0, 10):
    new_pairs = pairs.copy()
    for j in range(0, len(data.keys())):
        new_pairs = new_pairs + np.dot(np.transpose(pairs[j]), increments[j])
        char_counts = char_counts + np.dot(pairs[j], char_increments[j])
    pairs = new_pairs
char_counts = {list(chars)[i]:char_counts[i] for i in range(0, len(chars))}
print(char_counts)
print(f'Result: {max(char_counts.values())-min(char_counts.values())}')
```
